### Adding explaination to the data better understanding

```python
# Add the count of non null values

describe_data.loc['count', 'explaination'] = "This is the total non null values"
describe_data
```

```python
# Add the mean of the values 

describe_data.loc['mean', 'explaination'] = "Avg value to all the records"
describe_data
```

```python
# Add the std of the values

describe_data.loc['std', 'explaination'] = "Avg deviation of the record from the mean/Speard of the data"
describe_data
```

```python
# Add the min of the values

describe_data.loc['min', 'explaination'] = "Minimum value of the data/zeroth quartile"
describe_data
```

```python
# Add the 25% of the values

describe_data.loc['25%', 'explaination'] = "25% of data"
describe_data
```

```python
# Add the 50% of the values

describe_data.loc['50%', 'explaination'] = "50% of data"
describe_data
```

```python
# Add the 75% of the values

describe_data.loc['75%', 'explaination'] = "75% of data"
describe_data
```

```python
# Add the max of the values

describe_data.loc['max', 'explaination'] = "Maximum value of the data"
describe_data
```
